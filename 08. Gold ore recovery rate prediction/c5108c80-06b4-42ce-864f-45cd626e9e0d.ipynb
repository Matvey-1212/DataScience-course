**Расчёт эффективности** происходит по следующей формуле:

 $$recovery = \frac{{C \cdot (F - T)}}{{F \cdot (C - T)}} \times 100\%$$

- C — доля золота в концентрате после флотации/очистки;

- F — доля золота в сырье/концентрате до флотации/очистки;

- T — доля золота в отвальных хвостах после флотации/очистки.


```python
def recovery_formula(C, F, T):
    try:
        return (C * (F - T))/(F * (C - T)) * 100 
    except:
        return None
```

```python
def count_rougher_recovery(val): 
    rougher_C = val['rougher.output.concentrate_au']
    rougher_F = val['rougher.input.feed_au']
    rougher_T = val['rougher.output.tail_au']
    return recovery_formula(rougher_C, rougher_F, rougher_T)
```

```python
rougher_recovery = data_train.apply(count_rougher_recovery, axis=1)
```

```python
mae_recovery = mean_absolute_error(rougher_recovery, data_train['rougher.output.recovery'])
print(mae_recovery)
```

так как между некоторыми таймстемпами большие перерыви, заполнить их значенияями ближайших соседей будет некорректно. Для этого разделим наш датасет на части по непрерывным таймстемпам (перерыв более 3 часов) и в каждом заполним пропуски

```python
temp_datasets = []
hour = pd.Timedelta(hours=3)
start_i = 0
for i, diff in enumerate(data_all.reset_index()['date'].diff()):
    if diff > hour:
        temp_datasets.append(data_all.iloc[start_i:i].copy())
        start_i = i
temp_datasets.append(data_all.iloc[start_i:].copy())
print(f'количество временных промежутков: {len(temp_datasets)}')

count = 0
for el in temp_datasets:
    for i, diff in enumerate(el.reset_index()['date'].diff()):
        if diff > hour:
            count+=1
print(f'количество перерывов между таймствемпами внутри каждого временного промежутка: {count}')
```

```python
shape_of_df = []
for el in temp_datasets:
    shape_of_df.append(el.shape[0])
plt.hist(shape_of_df)
plt.title('Распределение размеров датасетов по временным промежуткам')
plt.show()
```

## Модель

зададим метрику sMAPe

$sMAPE= \frac{1}{N} \sum_{i=1}^{N} \frac {|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i}|)/2} \times 100\% $

$Final\hspace{0.2cm}sMAPE = 25\% \times sMAPE(rougher) + 75\% \times sMAPE(final)$

```python
def smape_error(y_true, y_pred):    
    smape = 0
    if y_true.shape[0] != y_pred.shape[0]:
        print('Error: not equal')
        return None
#     for i in range(len(y_true)):
#         smape += abs(y_true[i] - y_pred[i]) / ((abs(y_true[i]) + abs(y_pred[i])) / 2)
#     return smape / len(y_pred) * 100
    return np.mean(2* (np.abs(y_true - y_pred) / ((np.abs(y_true) + np.abs(y_pred))))) * 100

def print_smape(true, pred, mode=''):
    print(f'sMAPE{mode if mode == "" else " on "+mode}:    {smape_error(true,pred):.4f}')

    
def print_all_metrics(true, pred, mode=''):
    rmse = mean_squared_error(true,pred,squared = False)
    mae = mean_absolute_error(true,pred)
    r2 = r2_score(true,pred)
    smape = smape_error(true,pred)
    print()
    print(f'Metrics {mode+":" if mode == "" else " on "+mode +":"}')
    print("----------------------------------------------------------")
    print(f'RMSE _____: {rmse:.4f}')
    print(f'MAE ______: {mae:.4f}')
    print(f'R2 _______: {r2:.4f}')
    print(f'sMAPE ____: {smape:.4f}')
    print()

    
def smape(rougher_true, rougher_pred, final_true, final_pred):
    return 0.25 * smape_error(rougher_true, rougher_pred) + 0.75 * smape_error(final_true, final_pred)

```

Из-за сложности процесса очистки и недостаточной осведомленности в работе производства непонятны несколько моментов:
 - с чем связан отрицательный показатель уровня воды (особенность датчиков?)
 - с чем связан выброс и отрицательные значения объема воздуха (аномалия или опять же датчик?)
 - выбросы в краевых значения признаков recovery связаны с ошибкой на производстве, аномалией или же это особенность очистки золота, что зачастую итоговое содержание золота равно нулю?
 
 Лучшее понимание особенности производства позволит лучше изучить данные и учесть важные для анализа и обучения модели факторы.
